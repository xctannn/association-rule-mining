## Association Rule Mining 

```python
import mlxtend.frequent_patterns as fp
```

#### Question 1: Are there any differences in reading habits between different genders?

```python
# Finding frequent patttens
fp_sex_reading = fp.apriori(df_sex_reading_dummies,min_support=0.1,use_colnames=True)
fp_sex_reading['length'] = fp_sex_reading['itemsets'].apply(lambda x: len(x))
fp_sex_reading[(fp_sex_reading['length'] >= 2)].sort_values(by='support',ascending=False)
```

```python
rules_sex_reading = fp.association_rules(fp_sex_reading, metric="confidence", min_threshold=0.7)
rules_sex_reading['Kulczynski'] = rules_sex_reading.apply(lambda x: round(0.5 * x['support'] * ((1/x['antecedent support']) + (1/x['consequent support'])), 4),axis=1)
```

```python
rules_sex_reading = rules_sex_reading.sort_values(by='lift', ascending=False)
rules_sex_reading.head(10)
```

```python
rules_sex_reading.shape
```

**Interesting Rules**
<li>If a student studies less than 5 hours per week and reads scientific books or journals sometimes, it is quite likely the student is a male.
<li>A student is expected to read scientific books sometimes, when we know that the student is a male and read non-scientific books sometimes.
